```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Let us bring in the data
df_house_data = pd.read_csv("C:\\Users\\Dell\\Desktop\\Data Science\\house_prices_data.csv")
```

```python
#Let's see the first few rows of data
df_house_data.head(5)
```

```python
#Let's see the column and their data types
df_house_data.info()
```

```python
#Let's convert date column to year by extracting the year out of the date
df_house_data['date'] = df_house_data['date'].str[:4]
df_house_data['date'].head(5)
```

```python
#Let's convert the date column to integer so that we can obtain the age of the house from year it was built
df_house_data['date'] = df_house_data['date'].astype(int)
df_house_data['date'].dtypes
```

```python
#creating new column to capture age of building

df_house_data['house_Age'] = df_house_data['date']- df_house_data['yr_built']
df_house_data['house_Age'].head(5)
```

```python
#let's see the dataframe again and confirm column has been added
df_house_data.head(5)
```

```python
#Looking at things from a natuarl perspective, there are some variables that has nothing to do with the price of a house
#and there are some of them that are not really strong enough to be considered a good predictor of price from the data
#For example, at a glance, id here has nothing to do with price and will be removed. Then date is between 2014 and 2015
#so it does not offer much to be a good predictor here
#waterfront, view,lat,long,year renovated does not seem to offer much and will be removed
#Finally, year built will be removed because we have already detremined the age of the house

#drop some columns that may not be useful

df_house = df_house_data.drop(['id','date','waterfront','view','lat','long','yr_built','yr_renovated'],axis=1)
df_house.head(10)
```

```python
#let's convert zip codes to category in order not to confuse the model we are building
df_house['zipcode']= df_house['zipcode'].astype('category')
df_house.dtypes
```

```python
# let us see the summary statisctics of the data
df_house.describe()
```

```python
#from the above summary, some variables are of concern. For example, there are 33 bedrooms despite 75% of the 
#distribution being 4 bedrooms. There is also some concern with bathrooms, sqft_living among others having outliers
# that can skew our model
# Let's look at the differnt variables visually

boxplot = df_house.boxplot(column = ['bedrooms'])
```

```python
#The above plot shows the presence of outliers, we will have to remove it to minimize the negative effect on our model
#Let see bathroom
boxplot = df_house.boxplot(column = ['bathrooms'])
```

```python
boxplot = df_house.boxplot(column = ['sqft_living'])
```

```python
boxplot = df_house.boxplot(column = ['sqft_basement'])
```

```python
#We will try to remove the outliers and bridge the gap

index = df_house[(df_house['bedrooms'] >= 10)|(df_house['bedrooms'] <= 1)].index
df_house.drop(index, inplace=True)
df_house['bedrooms'].describe()
```

```python
index = df_house[(df_house['bathrooms'] >= 6)|(df_house['bathrooms'] <= 0)].index
df_house.drop(index, inplace=True)
df_house['bathrooms'].describe()
```

```python
index = df_house[(df_house['sqft_living'] >= 8000)|(df_house['sqft_living'] < 590)].index
df_house.drop(index, inplace=True)
df_house['sqft_living'].describe()
```

```python
#Let's look at the whole data again
df_house.describe()
```

```python
#Let us see how the features are correlated
house_corr = df_house.corr(method = 'pearson')
house_corr
```

```python
#Let us see the visuals
import seaborn as sns
sns.heatmap(house_corr,cmap="RdYlGn_r",linewidths=2.5)
plt.show()
```

```python

#We can see from the above that some variables are not correlated with price. For instance, age of house 
# has a weak negative correlatiob and this is actually not surprising

#Let us now seperate our data into features and target
X = df_house.drop('price',axis=1)
y= df_house[['price']]
```

```python
#Let us split the data into train,validate and test data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=1)
```

```python
#We further split the train dataset to get the validation dataset

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.05, random_state=1)
```

```python
#fitting multi linear regression to the dataset
#Natuarlly, we know that more than one variable affects the price of a house
#We will use the first 4 variables that show signs of correlation. we will start with bedrooms, bathrooms 
# sqft_living and grade
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train[['bedrooms','bathrooms','sqft_living','grade']],y_train)
```

```python
#Let's predict the test result

y_pred = regressor.predict(X_val[['bedrooms','bathrooms','sqft_living','grade']])
```

```python
#Let's validate the result and see if it works

from sklearn.metrics import r2_score
score = r2_score(y_val,y_pred)
```

```python
#This model is poor and we have to improve it
#The score is low, we will try and add more features to improve the accuracy
regressor = LinearRegression()
regressor.fit(X_train[['bedrooms','bathrooms','sqft_living','grade','condition','sqft_above','house_Age','sqft_basement','floors']],y_train)
```

```python
y_pred = regressor.predict(X_val[['bedrooms','bathrooms','sqft_living','grade','condition','sqft_above','house_Age','sqft_basement','floors']])
```

```python
#The model above is still weak. At this point, we have to add the zipcodes feature because it could hold some vital
#information to our model
#To add the zipcode, we need to encode to 0 and 1s
zipcode = pd.get_dummies(X['zipcode'],drop_first = True)
zipcode.head(5)
```

```python
#Let us drop the zipcode from the feature column as it is now redundant
X=X.drop('zipcode',axis=1)
```

```python
#Add the zipcode features to the main features
X = pd.concat([X,zipcode],axis=1)
```

```python
#Let's see what the new feature variables look like
X.describe()
```

```python
#Let's drop some additional variables that are not showing promise
X=X.drop(['sqft_lot','sqft_lot15','sqft_living15'],axis=1)
```

```python
#Let's distribute our data again
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=1)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.05, random_state=1)
```

```python
#Let's train our model with features including the zipcodes
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#let test with our test data
y_pred = regressor.predict(X_test)
```

```python
score = r2_score(y_test,y_pred)
```
